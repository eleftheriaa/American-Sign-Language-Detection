```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your CSV file
train_df = pd.read_csv("archive/sign_mnist_train.csv")
test_df = pd.read_csv("archive/sign_mnist_test.csv")


# View shape and first 5 rows
print(train_df.shape)
print(test_df.shape)

train_df.head()
```

```python
# Separate features and labels
X = train_df.drop('label', axis=1).values
y = train_df['label'].values
```

```python
# Reshape images into 28×28 matrices & Visualize a few examples

X_images = X.reshape(-1, 28, 28)
label_to_letter = {i: chr(65 + i) for i in range(26)}

plt.figure(figsize=(8,4))

for i in range(8):

    plt.subplot(2, 4, i+1)
    plt.imshow(X_images[i], cmap='gray')
    plt.title(f"Letter: {label_to_letter[y[i]]}")
    plt.axis('off')
plt.show()
```
