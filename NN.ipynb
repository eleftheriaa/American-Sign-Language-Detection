```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load Training Data
train = pd.read_csv("archive/sign_mnist_train.csv")
X_train = train.drop("label", axis=1).values
y_train = train["label"].values

# Load Test Data
test = pd.read_csv("archive/sign_mnist_test.csv")
X_test = test.drop("label", axis=1).values
y_test = test["label"].values

```

```python
from sklearn.preprocessing import StandardScaler

# normalize to [0,1]
X_train = X_train / 255.0
X_test = X_test / 255.0

# standardize each pixel feature
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# update label 0 to be label 9
y_train = np.where(y_train==0,9,y_train)
y_test = np.where(y_test==0,9,y_test)

# count unique classes in y_train / y_test
unique_classes = np.unique(y_train)
unique_classes_test = np.unique(y_test)

print("Unique classes in y_train:", unique_classes)
print("Unique classes in y_test:", unique_classes_test)

n_class = len(unique_classes)
print("Number of unique classes:", n_class)
```
